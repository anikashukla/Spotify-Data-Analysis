# **SPOTIFY DATA ANALYSIS** 






```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

# FIND TEN LEAST POPULAR SONGS 

```python
sorted_df = df_tracks.sort_values('popularity',ascending = True).head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

# TOP TEN POPULAR SONGS WITH POPULARITY > 90


```python
most_popular = df_tracks.query('popularity>90',inplace = False).sort_values('popularity', ascending= False)
most_popular[:10]
```

# MAKE THE RELEASE DATE COLUMN AS THE INDEX COLUMN

```python
df_tracks.set_index("release_date", inplace= True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

# FIND THE NAME OF THE ARTIST IN THE 46TH ROW OF THE DATASET


```python
df_tracks[["artists"]].iloc[46]
```

# CONVERT DURATION OF THE SONGS FROM MILLISECONDS TO SECONDS 

```python
df_tracks['duration']= df_tracks['duration_ms'].apply(lambda x: round(x/1000))
df_tracks.drop('duration_ms', inplace = True, axis=1)
```

```python
df_tracks.duration.head()
```

# CORRELATION MAP 

```python
corr_df= df_tracks.drop(["key",'mode','explicit'],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df,annot=True,fmt='.1g', vmin=-1, vmax=1, center=0, cmap='flare', linewidths=1, linecolor='Black')
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

# SAMPLING ONLY 5 PERCENT OF THE WHOLE DATASET


```python
sample_df = df_tracks.sample(int(0.005*len(df_tracks)))
print(len(sample_df))
```

# REGRESSION PLOT BETWEEN LOUDNESS AND ENERGY 

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y="loudness",x="energy", color = "c").set(title = "Loudness vs Energy Correlation")
```

# REGRESSION PLOT BETWEEN POPULARITY AND ACOUSTICNESS

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y="popularity",x="acousticness", color = "green").set(title = "Popularity vs Acousticness Correlation")
```


# DATA ANALYSIS BASED ON THE GENRE OF SONGS 

```python
df_genre = pd.read_csv('SpotifyFeatures.csv')
df_genre.head()
```

```python
plt.title("Duration  vof the songs in Different Genres")
sns.color_palette("rocket", as_cmap=True)
sns.barplot(y='genre', x='duration_ms', data = df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize = (10,5))
famous = df_genre.sort_values("popularity", ascending=False).head(10)
sns.barplot(y='genre', x='popularity', data=famous).set(title = 'Top 5 genres by Popularity')
```
